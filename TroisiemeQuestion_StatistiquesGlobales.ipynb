```python
import pandas as pd
fichier = pd.read_excel("AMD.xlsx")
fichier.drop(['Unnamed: 2','Unnamed: 4','Unnamed: 9'], axis=1, inplace=True)
fichier.dropna(inplace=True)
fichier
```
